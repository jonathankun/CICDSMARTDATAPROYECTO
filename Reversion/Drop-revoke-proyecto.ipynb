```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("nameContainer","unit-catalog")
dbutils.widgets.text("nameStorage","adlsproyectojjrr")
dbutils.widgets.text("catalogo","catalog_dev")
```

```sql
%python

nameContainer = dbutils.widgets.get("nameContainer")
nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta = f"abfss://{nameContainer}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.filmdetails;
DROP TABLE IF EXISTS ${catalogo}.bronze.movies;
DROP TABLE IF EXISTS ${catalogo}.bronze.posterpath;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta}/tablas/filmdetails", True)
dbutils.fs.rm(f"{ruta}/tablas/movies", True)
dbutils.fs.rm(f"{ruta}/tablas/posterpath", True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.movies_transformed;
```

```sql
%python
# REMOVE DATA (Silver)
dbutils.fs.rm(f"{ruta}/tablas/movies_transformed", True)
```
