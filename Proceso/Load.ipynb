```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_dev")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
%sql 
DROP TABLE IF EXISTS catalog_dev.golden.golden_movies_partitioned;
```

```python
    df_movies_transformed = spark.table(f"{catalogo}.{esquema_source}.movies_transformed")
```

```python
df_movies_transformed.show()
```

```python
df_transformed = df_movies_transformed.groupBy(col("anio")).agg(
                                                     count(col("titulo")).alias("conteopeliculas"),
                                                     max(col("duraciontotal_min")).alias("max_duracion_min"),
                                                     min(col("duraciontotal_min")).alias("min_duracion_min"),
                                                     avg(col("duraciontotal_min")).alias("avg_duracion_min"),
                                                     max(col("record_usuario")).alias("max_record_usuario"),
                                                     avg(col("record_usuario")).alias("avg_record_usuario"),
                                                     min(col("record_usuario")).alias("min_record_usuario"),
                                                     max(col("votos")).alias("max_votos"),
                                                     avg(col("votos")).alias("avg_votos"),
                                                     min(col("votos")).alias("min_votos"),
                                                     max(col("ingresos_usd")).alias("max_ingresos_usd"),
                                                     min(col("ingresos_usd")).alias("min_ingresos_usd"),
                                                     max(col("presupuesto_usd")).alias("max_presupuesto_usd"),
                                                     min(col("presupuesto_usd")).alias("min_presupuesto_usd"),
                                                     avg(col("presupuesto_usd")).alias("avg_presupuesto_usd"),
                                                     max(col("ganancia_usd")).alias("max_ganancia_usd"),
                                                     min(col("ganancia_usd")).alias("min_ganancia_usd"),
                                                     avg(col("ganancia_usd")).alias("avg_ganancia_usd"),
                                                     max(col("titulo")).alias("titulo"),
                                                     max(col("genero")).alias("genero"),
                                                     max(col("idioma")).alias("idioma"),
                                                     max(col("fecha_lanzamiento")).alias("fecha_lanzamiento"),
                                                     max(col("userscore_categoria")).alias("userscore_categoria"),
                                                     max(col("director")).alias("director"),
                                                     max(col("actores")).alias("actores"),
                                                     max(col("imagen_poster")).alias("imagen_poster"),
                                                     max(col("imagen_fondo")).alias("imagen_fondo"),
                                                     max(col("anios_lanzamiento")).alias("anios_lanzamiento")
                                                     ).orderBy(col("anio").desc())
```

```python
df_transformed.show()
```

```python
df_transformed.write.mode("overwrite").saveAsTable(f"{catalogo}.{esquema_sink}.golden_movies_partitioned")
```
